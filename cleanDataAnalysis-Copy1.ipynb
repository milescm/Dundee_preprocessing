```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('csvfile/cleanData.csv')
```

```python
df.drop(['Unnamed: 0'],axis=1, inplace=True)
```

```python
# 전류 소모값이 음수인 에러 데이터 확인
k = -1
for i in df['Total kWh']:
    k = k+1
    if(i < 0):
        print(k)
        
```

```python
# 전류 소모값이 음수인 에러 데이터 제거

df = df[df['Total kWh']>=0]
```

```python
# append하여 start time 기준으로 데이터 집어넣기

k = -1
startTime_11 = []

for i in df['Start Time']:
    k = k+1
    i = i.split(':')[0]
    if(i == '11'):
        print(df.iloc[k])
        startTime_11.append(df.iloc[k])
```

```python
pd_data11 = pd.DataFrame(startTime_11)
pd_data11.reset_index(inplace=True)
```
